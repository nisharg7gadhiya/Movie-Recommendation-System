```python
tfid = TfidfVectorizer()
tfid_matrix = tfid.fit_transform(df1['genres'])
```

```python
tfid_matrix
```

```python
similarity = linear_kernel(tfid_matrix, tfid_matrix)
```

```python
listing = pd.Series(df1.index, index = df1['title'])
index = df1['title']

def recommendation(obj):
  idx = listing[obj]
  sim_scores = list(enumerate(similarity[idx]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[1:6]
  movie_indices = [i[0] for i in sim_scores]
  return index.iloc[movie_indices]
```

```python
recommendation('Toy Story (1995)')
```
